```python
#Importing Relevant Libraries:
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython.display import display_html 
from functools import reduce

#Importing R&D Datasets:
df_expenditure_total = pd.read_csv('total-research-and-development-expenditure-annual.csv')
df_patents = pd.read_csv('patents-applied-awarded-and-owned-by-sector.csv')
df_expenditure_sector = pd.read_csv('research-and-development-expenditure-by-type-of-cost.csv')
df_grants = pd.read_excel('grant-calls-and-awarded-grant-projects-recipient-and-organisation.xlsx')
```

# How has R&D expenditure changed over the years?

```python
#Display DataFrame Details
print('*** Overview of Total R&D Expenditure Dataset  *** \n')
print(f'{df_expenditure_total.info()} \n')

#Oragnsing and Display DataFrame 
data_exp = []
df_exp = df_expenditure_total.loc[0:26]
last = df_expenditure_total.loc[26:29]
for x in np.split(df_exp, 3):
    data_exp.append(x)

year = df_expenditure_total['year']
exp = df_expenditure_total['rnd_expenditure']
    
print('*** Total R&D Expenditure from 1980 to 2018 *** ')   
exp1 = data_exp[0].style.set_table_attributes("style='display:inline'")
exp2 = data_exp[1].style.set_table_attributes("style='display:inline'")
exp3 = data_exp[2].style.set_table_attributes("style='display:inline'")
exp4 = last.style.set_table_attributes("style='display:inline'")
display_html(exp1._repr_html_() + exp2._repr_html_() + exp3._repr_html_() + exp4._repr_html_() , raw = True)

#Text-Based Analysis on R&D Expenditure
exp_min = df_expenditure_total['rnd_expenditure'].min()
exp_max = df_expenditure_total['rnd_expenditure'].max()
exp_change = round(((exp_max - exp_min)/exp_min) * 100)

print(f'Text based analysis from R&D expedniture data: \n')
print(f"From the data, the year that has seen the least R&D expenditure is in the year {year[df_expenditure_total['rnd_expenditure'].idxmin()]} at a cost of ${exp_min}")
print(f"From the data, the year that has seen the most R&D expenditure is in the year {year[df_expenditure_total['rnd_expenditure'].idxmax()]} at a cost of ${exp_max}\n")
print(f"The approximate total R&D expenditure across the years amounts to ${round(df_expenditure_total['rnd_expenditure'].sum())}")
print(f"The approximate average R&D expenditure across the years is ${round(df_expenditure_total['rnd_expenditure'].mean())}")
print(f"The approximate median R&D expenditure across the years is ${round(df_expenditure_total['rnd_expenditure'].median())}\n")
print(f"Additional Statistcs about R&D expenditure:")
display(df_expenditure_total.describe())

#Plotting Line Chart to show changes in R&D Expenditure over the years
fig, ax = plt.subplots(figsize = (16, 8))
plt.xlabel("Years", fontsize = 12)
plt.xticks(np.arange(year.min(), year.max() + 1, 2))
plt.ylabel("R&D Expenditure", fontsize = 12)
plt.grid(axis = 'y', alpha = 0.75)
plt.grid(axis = 'x', alpha = 0.75)
plt.title("Line Graph on Total R&D Expenditure over the Years", fontsize = 16)
plt.plot(year, exp, label = "Total R&D Expenditure" , linestyle = "solid", marker = "o",color = "purple")

#Annotation of % Increase
ax.text(0.29, 0.9, f'{exp_change}% increase in R&D Expenditure from 1990 to 2018',
     size = '15',
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes,
     bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})
plt.show()

#Analysis with Visulisation and Summary
print(f"\nAccording the line chart above, it is evident that there has been a steadily increasing trend in R&D expenditure, showing \nlittle signs of slowing down. The only oddity being in 2009 where there was a slight decrease in R&D expediture.")
print(f"\nDespite this from the years 1990 to 2018 there has been a whopping {exp_change}% increase in R&D Expenditure. In conclusion, R&D in\nSingapore is growing rapidly and will most likely continue to in the future. As Singapore is a small island it is important\nthat as a country it is cotinously trying innovate new ways deal with the lack of resources and implenet new technologies to\ngrow as a country.")
```

```python
#Display DataFrame Details - Patents
print('*** Overview of Patents Applied/Awarded Dataset  *** \n')
print(f'{df_patents.info()} \n')

#Oragnsing and Display DataFrame 
data_pat = []
for x in np.split(df_patents, 3):
    data_pat.append(x)

#Getting Dataset Infomation
year = df_patents['year']
sector = df_patents['sector']
pat_applied = df_patents['patents_applied']
pat_awarded = df_patents['patents_awarded']
pat_owned = df_patents['patents_owned']

pat_applied_min = df_patents['patents_applied'].min()
pat_applied_max = df_patents['patents_applied'].max()

pat_awarded_min = df_patents['patents_awarded'].min()
pat_awarded_max = df_patents['patents_awarded'].max()

pat_owned_min = df_patents['patents_owned'].min()
pat_owned_max = df_patents['patents_owned'].max()
    
print('*** Total Patents Awarded/Applied/Owned Across 2000 to 2018 *** ')   
pat1 = data_pat[0].style.set_table_attributes("style='display:inline'")
pat2 = data_pat[1].style.set_table_attributes("style='display:inline'")
pat3 = data_pat[2].style.set_table_attributes("style='display:inline'")
display_html(pat1._repr_html_() + pat2._repr_html_() + pat3._repr_html_(), raw = True)

print(f"Text Based Analyis from Patents Dataset:\n")
print(f"From the data, the year that has seen the least patents applied is the year {year[df_patents['patents_applied'].idxmin()]} with a total of {pat_applied_min} patents applied. ")
print(f"From the data, the year that has seen the most patents applied is the year {year[df_patents['patents_applied'].idxmax()]} with a total of {pat_applied_max} patents applied.")
print(f"The total patents applied over the years is {df_patents['patents_applied'].sum()}")
print(f"The mean amount patents applied over the years is {df_patents['patents_applied'].mean()}\n")

print(f"From the data, the year that has seen the least patents awarded is the year {year[df_patents['patents_awarded'].idxmin()]} with a total of {pat_awarded_min} patents awarded. ")
print(f"From the data, the year that has seen the most patents awarded is the year {year[df_patents['patents_awarded'].idxmax()]} with a total of {pat_awarded_max} patents awarded.")
print(f"The total patents awarded over the years is {df_patents['patents_awarded'].sum()}")
print(f"The mean amount patents awarded over the years is {df_patents['patents_awarded'].mean()}\n")

print(f"From the data, the year that has seen the least patents owned is the year {year[df_patents['patents_owned'].idxmin()]} with a total of {pat_owned_min} patents owned. ")
print(f"From the data, the year that has seen the most patents ownded is the year {year[df_patents['patents_owned'].idxmax()]} with a total of {pat_owned_max} patents owned.")
print(f"The total patents owned over the years is {df_patents['patents_owned'].sum()}")
print(f"The mean amount patents owner over the years is {round(df_patents['patents_owned'].mean())}\n")

#Extracting Infomation from Dataset - Public/Private and Total(Awarded Patents)
public_award = [] 
private_award = []
total_award = []

award = df_patents[df_patents['year'] > 2004]['patents_awarded']
#Getting Additional Patent Award Statistics
for i in range(2,30,2):
    #Public
    temp_pub = award[i]
    public_award.append(temp_pub)
    
    #Private
    temp_pri = award[i+1]
    private_award.append(temp_pri)
    
    #Total
    temp_total = temp_pub + temp_pri
    total_award.append(temp_total)  

#Storing infomation into DataFrame for Text Based Analysis

df_overall = {
    'Public': public_award,
    'Private': private_award,
    'Total': total_award
}

df_awarded = pd.DataFrame(df_overall)

#Calaculations on Private vs Public
total_awa = df_awarded['Total'].sum()
pub_total_awa = df_awarded['Public'].sum()
pri_total_awa = df_awarded['Private'].sum()
diff = pri_total_awa - pub_total_awa
percentage = round((diff/total_awa) *100)

df_describe = df_patents.describe()
pat1 = df_describe.style.set_table_attributes("style='display:inline'")
pat2 = (df_awarded.loc[0:7]).style.set_table_attributes("style='display:inline'")
pat3 = (df_awarded.loc[8:13]).style.set_table_attributes("style='display:inline'")
print("Additional Statistics about Patents Distributed Dataset:")
display_html(pat1._repr_html_() + " " + pat2._repr_html_() + pat3._repr_html_() , raw = True)

#Graph
labels = np.arange(year.min()+5, year.max()+1)
num = np.arange(0, 2000, 200)
x = np.arange(len(labels))  # the label locations
width = 0.50  # the width of the bars
fig, ax = plt.subplots(figsize = (20, 10))
plt.xlabel("Years", fontsize = 15)
plt.ylabel("Number of Patents", fontsize = 15)
plt.title("Bar Graph of Patents Stuff", fontsize = 18)
plt.grid(axis = 'y', alpha = 0.75)
plt.grid(axis = 'x', alpha = 0.75)
ax.set_xticks(x+0.25)
ax.set_xticklabels(labels, Fontsize = 15)
ax.set_yticklabels(num , Fontsize = 15)

fig.tight_layout()

X = np.arange(14)
plt.bar(X + 0.25, private_award, color = 'g', width = 0.25, edgecolor = 'black', linewidth = 1.2, label = "Private Awarded")
plt.bar(X + 0.00, public_award, color = 'r', width = 0.25, edgecolor = 'black', linewidth = 1.2, label = "Public Awarded")
plt.bar(X + 0.50, total_award, color = 'b', width = 0.25, edgecolor = 'black', linewidth = 1.2, label = "Total Awarded")
legend = plt.legend(loc = 'lower right', bbox_to_anchor = (1, -0.2), prop={'size': 15})
plt.show()

#Analysis with Visulisation and Summary
print("After cleaning the data, we can obtain the distribution of patents awarded based on Public and Private Sectors. Using the bar\nchart above as a visual aid, to summarise: \n")
print("Over the years the number of patents awarded as a whole has increased significantly, however much more awards are being\ndistributed to Private Sectors as compared to Public Sectors.")
print(f"In total over {total_awa} patents have been awarded, in which {pub_total_awa} patents are towards Public Sectors and {pri_total_awa} are towards Private\nSectors. There is noticeable difference of {diff} more patents awarded towards Private Sectors, meaning that Private Sectors make up approximately {percentage}% of the total.")
```

# What is the correlation between R&D Expenditure and Number of Patents Distributed? 

```python
#Extracting Infomation from Dataset - Public/Private and Total(Applied Patents)
public_applied = [] 
private_applied = []
total_applied = []

award = df_patents[df_patents['year'] > 2004]['patents_applied']

#Extracting Expenditure Infomation
expenditure = df_expenditure_total.loc[15:29]['rnd_expenditure']

#Getting Additional Patent Award Statistics
for i in range(2,30,2):
    #Public
    temp_pub = award[i]
    public_applied.append(temp_pub)
    
    #Private
    temp_pri = award[i+1]
    private_applied.append(temp_pri)
    
    #Total
    temp_total = temp_pub + temp_pri
    total_applied.append(temp_total)  

df_overall_app = {
    'Public': public_applied,
    'Private': private_applied,
    'Total': total_applied
}

df_applied = pd.DataFrame(df_overall_app)


print("*** Overall of Patent Distribution ***")
awa_d = (df_awarded.describe()).style.set_table_attributes("style='display:inline'")
app_d = (df_applied.describe()).style.set_table_attributes("style='display:inline'")
exp1 = df_expenditure_total.loc[15:29].style.set_table_attributes("style='display:inline'").set_caption('Expenditure')
pat1 = df_applied.style.set_table_attributes("style='display:inline'").set_caption('Patents Applied')
pat2 = df_awarded.style.set_table_attributes("style='display:inline'").set_caption('Patents Awarded')
display_html(pat1._repr_html_() + " " + pat2._repr_html_() + " " + exp1._repr_html_(), raw = True)


#Using Statisitcs for analysis and calculations
award = df_awarded['Total']
applied = df_applied['Total']

exp_min = (df_expenditure_total.loc[15:29]['rnd_expenditure']).min()
exp_max = (df_expenditure_total.loc[15:29]['rnd_expenditure']).max()
exp_percent = round(((exp_max - exp_min)/exp_min) * 100)

award_min = award.min()
award_max = award.max()
award_percent = round(((award_max - award_min)/award_min) * 100)

applied_min = award.min()
applied_max = award.max()
applied_percent = round(((applied_max - applied_min)/applied_min) * 100)


#Get Correlation Coefficent
r = np.corrcoef(expenditure, award)
r = r.flatten()
r_awa = round(r[1],2)

r_app = np.corrcoef(expenditure, applied)
r_app = r.flatten()
r_app = round(r[1],2)

#Text-Based Analysis
print(f"Since 2005 to 2018, R&D Expenditure has increased more then two folds with a {exp_percent}% increase in just 13 years. With this increase in R&D Expenditure, it would appear that number of patents being applied and awarded has also shot up.\n")
print(f"In fact, the number of Patents Awarded incresed from {award_min} in 2005 to {award_max} in 2018, which is a {award_percent}% increase in 13 years. For\nPatents Applied, it increased from {applied_min} in 2005 to {applied_max} in 2018, which is a {applied_percent}% increase over the 13 years.")

#Scatter Plot (Awarded and Applied Patents)
num = np.arange(800, 3000, 300)

fig, ax = plt.subplots(1, 2 , figsize = (16,8))
fig.suptitle("Relationship between R&D Expenditure and Patent Distribution",fontsize=20)

ax[0].set_title("Patents Awarded", fontsize = 15)
ax[0].scatter(expenditure, award, color='purple', s=100)
ax[0].plot(expenditure, np.poly1d(np.polyfit(expenditure, award, 1))(expenditure), 'r' , color='blue')
ax[0].set_yticks(num)
 
ax[1].set_title("Patents Applied", fontsize = 15)
ax[1].scatter(expenditure, applied, color='orange' , s=100, marker="D")
ax[1].plot(expenditure, np.poly1d(np.polyfit(expenditure, applied, 1))(expenditure), 'r', color='red')
ax[1].set_yticks(num)

#Annotation of coefficent, awarded
ax[0].text(0.5, 0.8, f'Correlation Coefficent of R&D Expenditure vs Patents Awarded: {r_awa} ',
     size = '10',
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax[0].transAxes,
     bbox={'facecolor':'purple', 'alpha':0.5, 'pad':10})

#Annotation of coefficent, applied
ax[1].text(0.5, 0.8, f'Correlation Coefficent of R&D Expenditure vs Patents Applied: {r_app} ',
     size = '10',
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax[1].transAxes,
     bbox={'facecolor':'orange', 'alpha':0.5, 'pad':10})

for ax in ax.flat:
    ax.set(xlabel='Expenditure', ylabel='Number of Patents')

plt.show()

#Conclusion
print(f"According the the scatter plots it is clear that there is a relatively strong positive linear relationship between R&D\nExpenditure and Patents being Applied/Awarded, with a correlation coefficent of {r_awa}. ")
print("\nIn conlusion, there is clear correlation between R&D Expenditure and Patent Distributions. As technology is cotinously being\npushed to the edge to more and more, it is only natural that it's complexity will also increase. As such more effort and\ntime needs to be poured in to development. To keep up with demand, more talents and resources are required, hence more overall expenditure toward R&D. As such more companies will also apply for more patents therfore increasing number of patents awarded. This increasing trend of patents highlights the growth of innovating technolgy and is a goood indicator for measuring\nSingapore's technological growth. ") 
```
